```python
# Import required libraries & dependencies
from pathlib import Path
import pandas as pd
import warnings

# Use new API: Historic_Crypto
from Historic_Crypto import HistoricalData
from Historic_Crypto import Cryptocurrencies
```

```python
# Ignore warnings
warnings.filterwarnings("ignore")
```

```python
# Create search variables
btc = 'BTC-USD'
eth = 'ETH-USD'
ada = 'ADA-USD'
timeframe = 86400
start_date = '2022-01-01-00-00'
end_date = '2023-01-31-00-00'
```

```python
# Create funtion to get data
def get_data(pair, timeframe, start_date, end_date):
    # Get all data
    data = HistoricalData(
    pair,
    timeframe,
    start_date,
    end_date).retrieve_data()
    # Slice out close column
    data = data[['close']]
    # Set index name
    data.index.names = ['Date']
    return data
```

```python
# Create function to calculate losses, gains, and percentages
def calc_info(high, low, min, max):
    loss = round(high - low, 2)
    gain = round(max - min, 2)
    pct_loss = round((loss / high) * 100, 2)
    pct_gain = round((gain / max) * 100, 2)
    return loss, gain, pct_loss, pct_gain
```

# Portfolio
---

```python
# Concatenate dataframes into all_2023_prices
all_2023_prices = pd.concat([btc_2023_df,
                              eth_2023_df,
                              ada_2023_df,
                              sp500_2023_df], axis="columns", join="inner")
```

```python
# Set column names
all_2023_prices.columns=['BTC','ETH','ADA', 'S&P500']
```

```python
# Calculate percent change and drop na's
all_2023_returns = all_2023_prices.pct_change().dropna()
```

```python
# Display head of all_2023_returns
all_2023_returns.describe()
```

```python
# Set initial investment amount to 10k
initial_investment = 10000
```

```python
# Set the weights of the investments
weights = [0.1, 0.1, 0.2, 0.6]
```

```python
# Calculate returns by weights
split_returns = all_2023_returns.dot(weights)
split_returns.head()
```

```python
# Calculate cumulative returns
portfolio_returns = (1 + split_returns).cumprod()
portfolio_cum_returns = (initial_investment * portfolio_returns)
```

```python
# Create basic plot of cumulative portfolio returns
portfolio_cum_returns.plot(figsize=(12,4),
                           title='Portfolio Performance 2023',
                           ylabel='Price In USD')
```

```python
# Portfolio Analysis
# ------------------------------
# Initial Investment = $10k
# Last Price = $12,072
# Gain = $2,072 (20%)
```

### Analysis
All of the four securities had losses in 2022. The crypto losses were more compared to the S&P 500, which is to be expected due to the volatility of the crypto market in general. Bitcoin fell to a three-year low of $15,760 falling $31,694 or 66.8%. Being the most significant market mover, Bitcoin's fall surely caused the major falls seen for other tokens as well. Ethereum dropped 74% and ADA dropped a whopping 84%!

All of the four securites had gains in 2023. Bitcoin has seen a 30% bounce, the largest of the bunch during the month of January with the others moving slightly behind, but also with upward trajectories. Ethereum and Cardano both saw jumps of 27% and the S&P 500 has gained 9%. These upward moves have stopped the downward slope of a bearish market into what looks to be a possible breakout and reversal into an upward trend.

Using the data mentioned above, we created a mock portfolio to determine if adding crypto to one's investment portfolio is a good idea for 2023. Our early findings suggest it's a great time for the addition. From an initial investment of $10,000 (60% S&P 500, 10% Bitcoin, 10% Ethereum, and 20% Cardano), there was an approximate 20% gain ($2,072) by the end of January 2023. While there are more factors to consider, no other factors speak like the actual movement of the market value itself, so we conclude that:

- The crypto market crash from all-time highs in 2021 seems to be over.
- Since market tumbles have slowed and may possibly reverse, now is a great time to add crypto to a portfolio.
- With proper risk management, the potential for crypto gains in 2023 could deliver a boost of approximately 20% - 30% to the client's portfolio.
