```python
# Import required libraries & dependencies
import pandas as pd
import hvplot.pandas
import warnings
from Historic_Crypto import HistoricalData
from Historic_Crypto import Cryptocurrencies
```

```python
# Ignore warnings
warnings.filterwarnings("ignore")
```

```python
# Create ticker search variables
xrp = 'BTC-USD'
ada = 'ETH-USD'
sol = 'SOL-USD'
dot = 'DOT-USD'
ltc = 'LTC-USD'
fil = 'FIL-USD'
eos = 'EOS-USD'
doge ='DOGE-USD'
link = 'LINK-USD'
matic = 'MATIC-USD'
```

```python
# Set the timeframe
timeframe = 86400
```

```python
# Set start and end dates
start_date = '2022-01-01-00-00'
end_date = '2023-01-31-00-00'
```

```python
# Create funtion to get data
def get_data(pair, timeframe, start_date, end_date):
    data = HistoricalData(
    pair,
    timeframe,
    start_date,
    end_date).retrieve_data()
    data = data[['close']]
    return data
```

```python
# Run function to get BTC data
xrp_df = get_data(xrp, timeframe, start_date, end_date)
xrp_df.describe()
```

```python
# Run function to get ETH data
ada_df = get_data(ada, timeframe, start_date, end_date)
```

```python
# Run function to get SOL data
sol_df = get_data(sol, timeframe, start_date, end_date)
```

```python
# Run function to get DOT data
dot_df = get_data(dot, timeframe, start_date, end_date)
```

```python
# Run function to get LTC data
ltc_df = get_data(ltc, timeframe, start_date, end_date)
```

```python
# Run function to get FIL data
fil_df = get_data(fil, timeframe, start_date, end_date)
```

```python
# Run function to get EOS data
eos_df = get_data(eos, timeframe, start_date, end_date)
```

```python
# Run function to get DOGE data
doge_df = get_data(doge, timeframe, start_date, end_date)
```

```python
# Run function to get LINK data
link_df = get_data(link, timeframe, start_date, end_date)
```

```python
# Run function to get MATIC data
matic_df = get_data(matic, timeframe, start_date, end_date)
```

```python
# Concatenate dataframes
comparison_data = pd.concat([ada_df, sol_df, dot_df, ltc_df, 
                            fil_df, eos_df, doge_df, link_df, matic_df], 
                            axis=1,
                            keys=['ADA', 'SOL', 'DOT', 'LTC',
                            'FIL', 'EOS', 'DOGE', 'LINK', 'MATIC'])
```

```python
# Create new dataframe for comparison data slices
comparison_df = pd.DataFrame()
```

```python
# Slice columns to comparison dataframe
comparison_df['ADA'] = comparison_data['ADA']['close']
comparison_df['SOL'] = comparison_data['SOL']['close']
comparison_df['DOT'] = comparison_data['DOT']['close']
comparison_df['LTC'] = comparison_data['LTC']['close']
comparison_df['FIL'] = comparison_data['FIL']['close']
comparison_df['EOS'] = comparison_data['EOS']['close']
comparison_df['DOGE'] = comparison_data['DOGE']['close']
comparison_df['LINK'] = comparison_data['LINK']['close']
comparison_df['MATIC'] = comparison_data['MATIC']['close']
```

```python
# Preview comparison dataframe
comparison_df.head()
```

```python
# Plot the comparison dataframe
comparison_df.hvplot(frame_width=600,
                    frame_height=300,
                    title="Crypto Heavyweights Daily 1-Year",
                    xlabel='Dates',
                    ylabel='Price In USD',
                    fontsize='12px',
                    rot=60)
```

```python
# Plot a rolling 7-day means of crypto closing prices
comparison_df.rolling(window=7).mean().hvplot(title="Crypto Daily Averages 2022-2023",
                                            xlabel='Dates',
                                            ylabel='Price In USD',
                                            rot=60)
```
