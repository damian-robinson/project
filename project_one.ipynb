```python
# Import Dependencies
import os
import requests
import pandas as pd
from datetime import date
from dotenv import load_dotenv
import alpaca_trade_api as alpaca_api
%matplotlib inline
```

```python
# Load .env environment variables
load_dotenv('/Users/d369/Documents/resources/research/fintech/projects/project_one/.env')
```

```python
# Set Alpaca API key and secret
alpaca_api_key = os.getenv("APCA_API_KEY")
alpaca_secret_key = os.getenv("APCA_SECRET_KEY")
```

```python
# Create the Alpaca API object
alpaca = alpaca_api.REST(
    alpaca_api_key,
    alpaca_secret_key,
    api_version="v2")
```

```python
# Set dates and format as ISO format
start_date = '2020-01-01'
end_date = '2023-01-10'
```

```python
# Set the tickers
tickers = ['BTC', 'ETH', 'SOL']
```

```python
# Set timeframe to one day ("1Day") for the Alpaca API
timeframe = '1Day'
```

```python
crypto_df = alpaca.get_bars(
    tickers,
    timeframe,
    start=start_date,
    end=end_date).df
```

```python
# Remove time from date index
crypto_df.index = crypto_df.index.date
```

```python
# Set the index label to 'date'
crypto_df.index.name = 'date'
```

```python
# Preview dataframe
crypto_df
```

```python
# Separate ticker data
btc_data = crypto_df[crypto_df['symbol']=='BTC'].drop('symbol', axis=1)
eth_data = crypto_df[crypto_df['symbol']=='ETH'].drop('symbol', axis=1)
sol_data = crypto_df[crypto_df['symbol']=='SOL'].drop('symbol', axis=1)
```

```python
# Concatenate ticker dataframes into new dataframe
clean_crypto_df = pd.concat([btc_data, eth_data, sol_data], axis=1, keys=['BTC','ETH', 'SOL'])
clean_crypto_df
```

```python
# Create and empty DataFrame for closing prices
crypto_closes = pd.DataFrame()
```

```python
# Get the closing prices from initial dataframe
crypto_closes['BTC'] = clean_crypto_df['BTC']['close']
crypto_closes['ETH'] = clean_crypto_df['ETH']['close']
crypto_closes['SOL'] = clean_crypto_df['SOL']['close']
```

```python
# Preview crypto closes dataframe
crypto_closes.info()
```

```python
crypto_daily = crypto_closes.pct_change().dropna()
```

```python
crypto_daily.plot(ylabel='Close', xlabel='Date', title="Crypto Daily", rot=90, figsize=(12, 6))
```
